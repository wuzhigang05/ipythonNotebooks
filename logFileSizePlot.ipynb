```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
pd.set_option('display.max_columns', 50)
```

```python
df = pd.read_csv("metaData", index_col=0)
df.shape
```

# now conver the file size into MB scale by dividing 1024 * 1024 from each hour of each day

```python
df = df.apply(lambda x: x/(1024*1024))
```

```python
tdf = df.transpose()
tdf.head(10)
```

# make a line plot to see how the (log) data size has changed over time

```python
totalSizePerDayDf = tdf.apply(sum, axis = 1)
totalSizePerDayDf
```

```python
pd.date_range('2013-03-22', periods = len(totalSizePerDayDf))
```

```python
pd.tseries.index.DatetimeIndex(totalSizePerDayDf.index)
```

```python
totalSizePerDaySeries = pd.Series(totalSizePerDayDf.tolist(), index=pd.tseries.index.DatetimeIndex(totalSizePerDayDf.index))
totalSizePerDaySeries
```

```python
matplotlib.rcParams.update({'font.size': 18, 'font.family': 'serif'})
fig, ax = plt.subplots(figsize=(12, 8))

totalSizePerDaySeries.plot(style='r-', lw=3, label='Assets', x_compat=True)
ax.set_title("Assets log size change over time")
#ax.set_xticklabels(pd.Index([i.strftime('%Y-%m-%d') for i in pd.date_range('2013-03-22', '2013-11-18', freq='M')]))
ax.set_ylabel("Log Size(MB)")
ax.set_xlabel("date")
ax.legend(loc=2)
```

```python
matplotlib.rcParams.update({'font.size': 12, 'font.family': 'sans'})
```

**Comments**: in above plot, we can see that the log file size increased significantly after 2013-10-08.
It used to be have a total size around 10 Gb per day. This number bounced to 50-85 Gb per day after 2013-10-08, 
which is comparable to the size of tracking log data size.

# next I am going to plot how the log file size change over 24 hours.

```python
#type(tdf)
#tdf.shape
#tdf['label'] = pd.Series(['A'] * len(tdf), index=tdf.index)
#tdf.head(10)
plt.figure(figsize=(16, 8))
ax = tdf.boxplot()
```

```python
df
sizePerHourDf = df.apply(np.mean, axis= 1)
sizePerHourDf = np.round(sizePerHourDf)
sizePerHourDf
```

```python
pd.date_range(start = 0, end = 23, freq= 'H')
#pd.tseries.index.DatetimeIndex(sizePerHourDf.index)
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
sizePerHourDf.plot(style='r-', lw=3)
ax.set_title("log size change within a day")
ax.set_xticks(sizePerHourDf.index)
ax.set_xlim((0, 23))
ax.set_ylabel("log size (MB)")
ax.set_xlabel("hour(h)")
```

**comment**: Above graph actually indicates how many people were watching our video in a daily window.
We can see that the time period  7-9am is when our video websites have fewest visitors. 
The number of people watching our video gradually begin to increase from noon (10:00am) reach its 
maximum during period 19:00-1:00.
