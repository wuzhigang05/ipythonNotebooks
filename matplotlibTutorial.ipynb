```python
x = np.linspace(0, 5, 10)
x
```

```python
y = x ** 2
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0.1, 0.1, 0.9, 0.9])
ax1.plot(x, y, 'r')
ax1.set_title("title")
ax1.set_xlabel("x")
ax1.set_ylabel('y')
ax2 = fig.add_axes([0.2, 0.4, 0.4, 0.4])
ax2.plot(x, y, 'b')
```

```python
fig, axes = plt.subplots(nrows=1, ncols = 2)
for i, ax in enumerate(axes):
    ax.plot(x, y, 'ro-')
    ax.set_xlabel('x %d' % i)
    ax.set_ylabel('y %d' % i)
    ax.set_title("title %d" % i)
    
```

```python
fig, axes = subplots(figsize=(12, 8))
axes.spines['top'].set_color('blue')
axes.spines['bottom'].set_color('blue')
axes.spines['left'].set_color('red')
axes.spines['left'].set_linewidth(2)
axes.spines['right'].set_color('None')
axes.yaxis.tick_left()
axes.xaxis.tick_bottom()
for i, a in enumerate(axes.get_xticklabels()):
    if i != 0:
        a.set_color('blue')
        a.set_fontsize(18)
    else:
        a.set_color('None')
        a.set_fontsize(0)
        
for i in axes.get_yticklabels():
    i.set_color('red')
    i.set_fontsize(18)
```
